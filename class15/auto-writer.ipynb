```python
# asked chatgpt to write me a function to turn a string date to ap style

from datetime import datetime

def to_ap_style(date_str):
    """
    Convert a date string in common formats (M/D/YY, M/D/YYYY, YYYY-MM-DD)
    into AP Style (e.g. 'Aug. 27, 2025').
    """
    # Try several common input formats
    formats = ["%m/%d/%y", "%m/%d/%Y", "%Y-%m-%d"]
    for fmt in formats:
        try:
            dt = datetime.strptime(date_str, fmt)
            break
        except ValueError:
            pass
    else:
        raise ValueError("Unsupported date format.")

    # AP Style month abbreviations
    ap_months = {
        1: "Jan.", 2: "Feb.", 3: "March", 4: "April", 5: "May", 6: "June",
        7: "July", 8: "Aug.", 9: "Sept.", 10: "Oct.", 11: "Nov.", 12: "Dec."
    }

    month = ap_months[dt.month]
    day = dt.day
    year = dt.year

    return f"{month} {day}, {year}"
```

```python
print(f"The number one song for the week of {to_ap_style(num_1_song['Date'])} is {num_1_song['Song']}\
 by {num_1_song['Artist']}. It's been on the charts\
 for {num_1_song['Weeks in Charts']} weeks and moved\
 up {1-num_1_song['Last Week']} spots from last week.")
```
