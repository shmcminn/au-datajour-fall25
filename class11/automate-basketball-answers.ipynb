## ‚ûï Combining for each opponent

So now we can add up the total AU points and total opponent points for each opponent. For an opponent like "Navy" we should see totals added together, since we played them 3 times.

```python
scores_combined_df = scores_df.groupby("opp_name").sum()[['AU_score','opp_score']]
```

`scores_combined_df = ` creates a brand new dataframe (think of it like a new pivot table or new tab in a spreadsheet)

`scores_df.groupby("opp_name")` groups the scores_df dataframe into one row per value in `opp_name` column (one row per opponent) 

`.sum()` adds up the grouped values in every column

`[['AU_score','opp_score']]` only returns the AU_score and opponent score columns

```python
scores_combined_df
```

But wait! Something loks wrong. Did AU really score 586,074 points against Navy? 

Nope. 

This is "summing" *strings*, not integers, together. So 58 AU's first game, 60 from AU's second game and 74 from AU's third game ...

We need to tell Pandas that that column are actually *integers*. Not strings of text.

So we here we go back to our original scores_df dataframe and recast certain columns as integers (ignore the warnings that pop up.)

```python
scores_df['AU_score'] = scores_df['AU_score'].astype(int)
```

üßë‚Äçüíª **Do the same for opp_score**

```python
scores_df['opp_score'] = scores_df['opp_score'].astype(int)
```

Now we'll go back and create our groupby dataframe again, and this time it should know those 2 columns are integers that need to be summed

```python
scores_combined_df = scores_df.groupby("opp_name").sum()[['AU_score','opp_score']]
scores_combined_df
```

Better! 

So now we have the total for and against each opponent, but we really want the point differential. 

Just like we used division when we calculated points per game earlier on, we'll do the same here by creating a new column that subtracts opponent score from AU score. 

üßë‚Äçüíª **Create a new column in `scores_combined_df` called `point_diff` that is equal to `AU_score` - `opp_score`**

```python
scores_combined_df['points_diff'] = scores_combined_df['AU_score'] - scores_combined_df['opp_score']
```

```python
scores_combined_df
```

Last thing, let's sort the dataframe by that new column to see who AU did best and worst against overall

## üßë‚Äçü§ù‚Äçüßë Combining all that 

What we've done so far is 

* Get the 2024 table from the AU website
* Pull out the scores
* Standardize the opponents
* Drop the non-game total rows
* Combine the scores for each opponent

We've done this across several cells, but all in one cell it could look like this (again, ignore the warnings):

```python
import pandas as pd

# get the data from AU
web_tables = pd.read_html("https://aueagles.com/sports/mens-basketball/stats/2024-25")
scores_df = web_tables[5]

# pull out the scores
scores_df['AU_score'] = scores_df['Score'].str.split("-",expand=True)[0]
scores_df['opp_score'] = scores_df['Score'].str.split("-",expand=True)[1]

# standardize opponent name by removing vs or at
scores_df['opp_name'] = scores_df['Opponent'].str[3:]

# drop the non game total rows
scores_df = scores_df[scores_df['Date'].isna()==False]

# combine for each opponent
scores_df['AU_score'] = scores_df['AU_score'].astype(int)
scores_df['opp_score'] = scores_df['opp_score'].astype(int)
scores_combined_df = scores_df.groupby("opp_name").sum()[['AU_score','opp_score']]
scores_combined_df['points_diff'] = scores_combined_df['AU_score'] - scores_combined_df['opp_score']

# output the sorted combined scores dataframe
scores_combined_df.sort_values("points_diff")
```

Note the comments there, which aren't telling Python to do anything but just makes it more readable for us humans

Now we want to look at how school names are written, since we can't do a `groupby` if they're different.

Right off the bat we can see issues. Albany is in there as "ALBANY" and "Albany (NY)". Boston University looks like it's in there 3 different ways.

----------

To account for this, we can set up a *dictionary* that tells Pandas how we want each to be referred to. A dictionary is like a list, except instead of referring to values in it by an index ([0], [1], etc...), we refer to them by a *key*. 

So `sean_dict = {"hair": "brown", "name": "sean", "eyes": "blue"}` could be a dictionary representing me. 

If we output `sean_dict['hair']`, we would expect it to return "brown"

---------------------

üßë‚Äçüíª Here, we'll start a dictionary that says how we want to refer to the ones that need to change. **We'll go through and fill it in together.**

```python
school_names_dict = {
    'Albany (NY)': "ALBANY",
    'ARMY': "Army",
    'Army West Point': "Army"
}
```

Once we have that done, we'll want to make that a new column called `opp_name_cleaned`. We can change the names in `opp_name` column that are keys in our dictionary. 

```python
def replace_name(x):
    if x in school_names_dict:
        return school_names_dict[x]
    else:
        return x

giant_score_df_filtered['opp_name_cleaned'] = giant_score_df_filtered['opp_name'].apply(replace_name)
```

`def replace_name(x)` is a *function*. It takes an input (x) and then returns an output depending on what x is.

`if x in school_names_dict:` says if the input `x` exists as a key in our dictionary, do the following:

`return school_names_dict[x]` would in that case return the *value* of that dictionary key `x`

`else:` says if `x` does NOT exist as a key in our dictionary, do the following:

`return x` just returns the input `x` without doing anything to it. 

In total, this function either substitutes in a "clean" name or just returns the original name. But this above function doesn't "automatically" run. We have to tell Python or Pandas how to use it below:

`giant_score_df_filtered['opp_name_cleaned'] =` sets new column called `opp_name_cleaned`

`giant_score_df_filtered['opp_name']` accesses our old column called `opp_name`

`.apply(replace_name)` applies our `replace_name` function to that column 



-------------

## üßë‚Äçü§ù‚Äçüßëüóì Putting it all together, with all the years of data 

As a reminder, here's the old code block we wrote

```python
import pandas as pd

# get the data from AU
web_tables = pd.read_html("https://aueagles.com/sports/mens-basketball/stats/2024-25")
scores_df = web_tables[5]

# pull out the scores
scores_df['AU_score'] = scores_df['Score'].str.split("-",expand=True)[0]
scores_df['opp_score'] = scores_df['Score'].str.split("-",expand=True)[1]

# standardize opponent name by removing vs or at
scores_df['opp_name'] = scores_df['Opponent'].str[3:]

# drop the non game total rows
scores_df = scores_df[scores_df['Date'].isna()==False]

# combine for each opponent
scores_df['AU_score'] = scores_df['AU_score'].astype(int)
scores_df['opp_score'] = scores_df['opp_score'].astype(int)
scores_combined_df = scores_df.groupby("opp_name").sum()[['AU_score','opp_score']]
scores_combined_df['points_diff'] = scores_combined_df['AU_score'] - scores_combined_df['opp_score']

# output the sorted combined scores dataframe
scores_combined_df.sort_values("points_diff")
```

* changed variable names to match our giant_score_df instead of just one year at a time

(Note this will take a little bit to run as it will "re-scrape" each of those URLs)

```python
import pandas as pd

# *NEW* Scrape all the years from AU
years_list = ["2009-10", "2010-11", "2011-12", "2012-13", "2013-14", "2015-16", "2016-17", "2017-18", "2018-19", "2019-20", "2020-21", "2021-22", "2022-23", "2023-24", '2024-25']
giant_score_df = pd.DataFrame()

for year_string in years_list:
    web_tables = pd.read_html("https://aueagles.com/sports/mens-basketball/stats/" + year_string)
    scores_df = web_tables[5]
    giant_score_df = pd.concat([giant_score_df,scores_df])
    print(year_string, 'completed')
    time.sleep(4)

# standardize opponent name by removing vs or at (*moved up before we fix school names*)
giant_score_df['opp_name'] = giant_score_df['Opponent'].str[3:]

# drop the non game total rows (*moved up before we fix school names*)
giant_score_df_filtered = giant_score_df[giant_score_df['Date'].isna()==False]

# *NEW* fix school names that are written differently
school_names_dict = {
    'Albany (NY)': "ALBANY",
    'ARMY': "Army",
    'Army West Point': "Army"
}
giant_score_df_filtered['opp_name_cleaned'] = giant_score_df_filtered['opp_name'].apply(lambda x: school_names_dict[x] if x in school_names_dict else x)

# *NEW* set a minimum threshold of how many times AU has to play a team to be counted
counts = giant_score_df_filtered['opp_name_cleaned'].value_counts()
min_threshold = 3
giant_score_df_filtered_cts = giant_score_df_filtered[
    giant_score_df_filtered["opp_name_cleaned"].map(counts) >= min_threshold
]

# *NEW* from here on out we are referring to giant_score_df_filtered_cts to get all the scores, which replaces scores_df

# pull out the scores
giant_score_df_filtered_cts['AU_score'] = giant_score_df_filtered_cts['Score'].str.split("-",expand=True)[0]
giant_score_df_filtered_cts['opp_score'] = giant_score_df_filtered_cts['Score'].str.split("-",expand=True)[1]

# combine for each opponent
giant_score_df_filtered_cts['AU_score'] = giant_score_df_filtered_cts['AU_score'].astype(int)
giant_score_df_filtered_cts['opp_score'] = giant_score_df_filtered_cts['opp_score'].astype(int)
scores_combined_df = giant_score_df_filtered_cts.groupby("opp_name_cleaned").sum()[['AU_score','opp_score']] # *NEW* refers to opp_name_cleaned instead of opp_name
scores_combined_df['points_diff'] = scores_combined_df['AU_score'] - scores_combined_df['opp_score']

# output the sorted combined scores dataframe
scores_combined_df.sort_values("points_diff")
```

This table above reflects the total points for and against teams that AU played at least three times between the 2010 and 2025 seasons (excluding the 2015 season, which had no data).
